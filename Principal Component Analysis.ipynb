```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#getting dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer= load_breast_cancer() #  builtin datasets of scikit learn
```

```python
type(cancer)
```

```python
#its a dictionary
```

```python
cancer.keys()
```

```python
print(cancer['DESCR']) #30 attributes and 569 records
```

```python
#to predict whether is malign or benign
```

```python
#our task is to remove uneeded features or columns here
```

```python
df=pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head() #without target
```

```python
cancer['target']
```

```python
#pca is done before putting machine learning algos
```

```python
#finding first 2 pricipal components out of 30
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#each feature has single unit variance
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data= scaler.transform(df)
```

```python
#PCA
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
#transform data to first principal component
```

```python
x_pca= pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
#reduced 30 dimensions to 2
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#two components decided the cancer type easily, its a compression technique basically
```

```python
#they dont relate to one to one feature but set of features
```

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp
```

```python
sns.heatmap(df_comp, cmap='plasma')
```
