```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Classified Data', index_col=0)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler #fitting data and standardizing it
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS', axis=1)) #scale all the columns on similar grounds to avoid Curse of dimensionality
```

```python
scaled_features
```

```python
#Recreate df for features using scaled_features to have better understanding
df_feat=pd.DataFrame(scaled_features, columns=df.columns[:-1])
```

```python
df_feat.head()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X= df_feat
y=df['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#pick k=1
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
```

```python
#evaluation for 1 neighbour
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```

```python
error_rate=[] #pick multiple k values and check best k value
```

```python
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize=(10,6))
```

```python
error_rate
plt.plot(range(1,40), error_rate, color='blue', linestyle="dashed",
        marker='o', markerfacecolor='red', markersize=10)
```

```python
#pick not very high k value for small error , here go for 15
```

```python
knn=KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train, y_train)
pred=knn.predict(X_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```
